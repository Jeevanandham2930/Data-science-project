```python
# for speech-to-text
import speech_recognition as sr
from gtts import gTTS
import os
import pyjokes
import time
# for data
import datetime
import numpy as np
import pyttsx3
import wikipedia
import webbrowser
```

```python
class ChatBot():
    def __init__(self,name):
        print("----- starting up", name, "-----")
        self.name = name
        
    def speech_to_text(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as mic:
            print("listening...")
            audio = recognizer.listen(mic)
        try:
            self.text = recognizer.recognize_google(audio)
            print("Human --> ", self.text)
        except sr.UnknownValueError:
            print("Human -->  ERROR")

    def wake_up(self):
        print("Did you Say:", self.text)
        return True if self.name.lower() in self.text.lower() else False

    @staticmethod
    def text_to_speech(text):
        speed = 0.9
        voice=pyttsx3.init()
        rate = voice.getProperty('rate')
        voice.setProperty('rate', rate * speed)
        voice.say(text)
        voice.runAndWait()
        
        #print("AI --> ", text)
        #speaker = gTTS(text=text, lang="en", slow=False)
        #speaker.save("res.mp3")
        #os.system("start res.mp3")  # For Windows. For other OS, adjust the command
        #os.remove("res.mp3")
        
    def action_time(self):
        return datetime.datetime.now().time().strftime('%H:%M')
    def wishMe(self):
        hour = int(datetime.datetime.now().hour)
        if hour>= 0 and hour<12:
            print("Good Morning !")
            ai.text_to_speech("Good Morning !")
  
        elif hour>= 12 and hour<18:
            print("Good Afternoon !")
            ai.text_to_speech("Good Afternoon !")   
  
        else:
            print("Good Evening !")
            ai.text_to_speech("Good Evening !")
    
    def job_section(self):
        lis=np.random.choice(['Data Science','Full stack','Cloud','Devops','IOT'])
        return lis
        
    def open_youtube(self):
        return webbrowser.open("https://www.youtube.com")
    
    def open_google(self):
        return webbrowser.open("https://www.google.com")
    
    def open_instagram(self):
        return webbrowser.open("https://www.instagram.com")
    def songs(self):
        music_dir = "D:\MUSIC"
        songs = np.random.choice(os.listdir(music_dir))
        print('For you--->',songs)    
        random = os.startfile(os.path.join(music_dir,songs))
        return random
    
    def wiki(self):
        #wikipedia.set_lang("en")
        #inp=input('ENter')
        #result=wikipedia.summary(self.text,sentences=3)
        try:
            wikipedia.set_lang("en")
            result=wikipedia.summary(self.text,sentences=3)
            return result
        except wikipedia.exceptions.DisambiguationError as e:
            # Handle disambiguation error
            print(f"Ambiguous term. Options: {e.options}")
        except wikipedia.exceptions.PageError as e:
            # Handle page not found error
            print(f"Page not found")
        #return result
```

```python
#while True:
def run_command():
    ai = ChatBot(name="Siri")
    while True:
        try:
            ai.speech_to_text()
            if ai.wake_up():
                res = "Hello, I am Siri the AI. What can I do for you?"
                ai.wishMe()
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "time" in ai.text:
                res = ai.action_time()
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "what is your name" in ai.text:
                res= (f'My name {ai.name}')
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "how are you" in ai.text:
                res=('I am fine and how are you....?')
                print('AI-->',res)
                ai.text_to_speech(res)
            elif any(i in ai.text for i in ["who made you","who created you"]):
                res=('Developed by Jeevanandham..')
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "job" in ai.text:
                res= ai.job_section()
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "open YouTube" in ai.text:
                ai.open_youtube()
                print('AI-->',res)
                ai.text_to_speech(res)
                res='Opening YouTube in your web browser.' 
            elif any(i in ai.text for i in ["play songs","play music"]):
                res=ai.songs()
                print('AI-->',res)
                ai.text_to_speech(res)
            elif 'say joke' in ai.text:
                res=pyjokes.get_joke()
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "open Instagram" in ai.text:
                ai.open_instagram()
                res='Opening Instagram in your web browser.'
                print('AI-->',res)
                ai.text_to_speech(res)
            elif "open Google" in ai.text:
                ai.open_google()
                res='Opening Google in your web browser.'
                print('AI-->',res)
                ai.text_to_speech(res)
            elif any(i in ai.text for i in ['who','where','what','which','how much']):
                res=ai.wiki()
                print('AI-->',res)
                ai.text_to_speech(res)
            elif any(i in ai.text for i in ["thank","thanks"]):
                res = np.random.choice(["you're welcome!","I'm here if you need me!"])
                print('AI-->',res)
                ai.text_to_speech(res)
            elif any(i in ai.text for i in ["exit","close"]):
                res = np.random.choice(["Tata Bye","Have a good day","Goodbye","Hope to meet soon"])
                print('AI--->',res)
                ai.text_to_speech(res)
                break
        except:
            ai.text_to_speech('Waiting for your response')
    return True
```

```python
if __name__ == "__main__":
    run_command()
```
